```python
#形態素解析結果（neko.txt.mecab）を読み込むプログラムを実装せよ．ただし，各形態素は
#表層形（surface），基本形（base），品詞（pos），品詞細分類1（pos1）をキーとするマッピング型に格納し
#1文を形態素（マッピング型）のリストとして表現せよ．第4章の残りの問題では，ここで作ったプログラムを活用せよ．
import codecs


with codecs.open("neko.txt.mecab",'r','utf-8') as f:
        data = f.readlines()
mecab_list=[]
temp_dict ={}

for temp_word in data:
        temp_word = temp_word.replace('\t', ',')
        temp_word = temp_word.replace('\n', '')
        if(temp_word.count(',')==9 or temp_word.count(',')==7):
            temp_list = temp_word.split(',')
            temp_dict={'surface':temp_list[0],'base':temp_list[7],'pos':temp_list[1],'pos1':temp_list[2]}
            mecab_list.append(temp_dict)
        else:
            continue
print(mecab_list[:5])
```

```python
#2つの名詞が「の」で連結されている名詞句を抽出せよ
import codecs
import ast
with codecs.open("neko.txt.mecab",'r','utf-8') as f:
        data = f.readlines()
words_list=[]
temp_w=''
flag=0
for temp_word in data:
        temp_word = temp_word.replace('\t', ',')
        temp_word = temp_word.replace('\n', '')
        if(temp_word.count(',')==9 or temp_word.count(',')==7):
            temp_list = temp_word.split(',')
            temp_dict={'surface':temp_list[0],'base':temp_list[7],'pos':temp_list[1],'pos1':temp_list[2]}
            if(temp_list[1]=='名詞'and flag==0):
                temp_w=temp_list[0]
                flag=1
                
            elif(temp_list[0]=='の'and flag==1):      
                temp_w+=temp_list[0]
                flag=2
                
            elif(temp_list[1]=='名詞'and flag==2):
                temp_w+=temp_list[0]
                words_list.append(temp_w)
                flag=0
                
        else:
            continue
print(words_list)
```
