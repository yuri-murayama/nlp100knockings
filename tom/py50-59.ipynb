```python
#50 . : ; ? !を判定基準に文を分ける

import re

#文章を返す関数を作る
def f():
    with open ("nlp.txt",encoding='utf-8') as file:
        for line in file:
            #区切り文字を開業に書き換える
            nl=re.sub(r'((\.|;|:|\?|!)(\s+)([A-Z]))',r'\n',line)
            #改行で区切る
            sentences=re.split(r'\n',nl)
            for sentence in filter(lambda w:len(w)>0,sentences):
            #文章返す
                yield sentence
#新しいファイルを作る        
with open ("text.txt",'w',encoding='utf-8')as file2:
    for sentence in f():
        #一文ずつ書いていく
        file2.write(sentence+'\n')
    
    
```

```python
#51空白を単語の区切りとみなし，50の出力を入力として受け取り，1行1単語の形式で出力せよ．ただし，文の終端では空行を出力せよ．
import re
def f():
    with open("text.txt",encoding='utf-8')as file:
        for line in file :
            temp=re.sub(r'\.|,|"|\?|\(|\)','',line)
            words=temp.split(" ")
            for word in filter(lambda w:len(w)>0,words):
                yield word


with open ("text2.txt","w",encoding='utf-8')as file2:
    for word in f():
        file2.write(word+'\n')
```

```python
#52.51の出力を入力として受け取り，Porterのステミングアルゴリズムを適用し，単語と語幹をタブ区切り形式で出力せよ． 
#Pythonでは，Porterのステミングアルゴリズムの実装としてstemmingモジュールを利用するとよい．
from stemming.porter2 import stem
import re

def f():
    with open("text2.txt",encoding='utf-8')as file:
        for line in file :
            line=re.sub(r'\n','',line)
            wstem=stem(line)
            word_stem=line+'\t'+wstem
            yield word_stem


with open ("text3.txt","w",encoding='utf-8')as file2:
    for word_stem in f():
        file2.write(word_stem+'\n')

```
